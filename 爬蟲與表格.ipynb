# 書的表格
1. make a csv about books' basic inforamtion
2. must have columns = ['title', 'Product Type, 'price', 'description']

```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
res = requests.get('http://books.toscrape.com/')
soup = BeautifulSoup(res.text)
```

```python
soup
```

```python
list_url_books = ['https://books.toscrape.com/' + i['href'] for i in soup.select('h3 a')]
```

```python
list_url_books
```

```python
import csv
```

```python
def save_to_csv(items,file):
  with open(file,'w+',newline='',encoding='utf-8') as fp:
    writer = csv.writer(fp)
    for item in items:
      writer.writerow(item)
```

```python
if res.status_code == requests.codes.ok:
  books = [['title', 'Product Type', 'price', 'description']]
  for j in list_url_books:
    res_2 = requests.get(j)
    soup_2 = BeautifulSoup(res_2.text)
    title = soup_2.select_one('div h1').text
    description = soup_2.select('p')[3].text
    producttype = soup_2.select('td')[1].text
    pric = soup_2.select('td')[2].text
    book = [title.strip(),
            producttype.strip(),
            pric.strip(),
            description.strip()]
    books.append(book)
  for book in books:
    print(book)
  save_to_csv(books,'books.csv')
else:
  print('error')
```
